```python
# NIVELL 1 - EXERCICI 1: Importa com un DataFrame l'arxiu sprint10.xlsx. 
# Assegura't que el fitxer s'importa correctament, amb els noms de columnes que li corresponen, sense manipular l'arxiu original.

# Primer, importem la llibreria pandas
import pandas as pd

df = pd.read_excel(r"C:\Users\joanr\OneDrive\Escritorio\FAD\sprint10.xlsx")

# Eliminem les dues primeres files (0 i 1)
df = df.drop(index=[0, 1]).reset_index(drop=True)

# Assignem la primera fila com a nova capçalera (columnes)
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)

# Eliminem les dues primeres columnes
df = df.drop(df.columns[[0]], axis=1)

# Ordenem el DataFrame pel país d'origen. En cas d'empat, ordena pel nom de la ciutat.
df.sort_values(["País d'origen", "Ciutat"], ascending=(True, True), inplace= True)
# Mostrem les primeres 10 columnes
df.head(10)
```

```python
# Addicionalment, fes un print on comprovi que el DNI només té valors únics.

# Comptem el nombre de valors únics, comprovant que es correspon amb el nombre de value_counts
numero_unicos = df["DNI"].nunique()
print(f"Número de valores únicos en DNI: {numero_unicos}")
```

```python
# EXERCICI 2: 
import numpy as np
# Crea una columna que sigui el nom complet.
df['Nom_Cognoms'] = df[["Nom", "Cognoms"]].agg(" ".join, axis=1) # concatenacion strings es mes facil

# Crea una columna si la persona és nascuda a Espanya o no.
df["Espanyol"] = df["País d'origen"].apply(lambda x: "Sí" if x == "Espanya" else "No")

# Posa el DNI com a índex del DataFrame (noms de files).
df.set_index("DNI", inplace= True) 

# Substitueix el nom de les columnes Dia de Naixement, Mes de Naixement i Any de Naixement per Dia, Mes i Any.
df.rename(columns={"Dia de Naixement": "Dia", "Mes de Naixement": "Mes", "Any de Naixement": "Any"}, inplace= True)

# Substitueix H per Home, D per Dona, A per Altres i NC per una dada faltant (nan/null/na).
df["Gènere"] = df["Gènere"].replace({"A": "Altres", "H": "Home", "D": "Dona", "NC": np.nan})
df
```

```python
# EXERCICI 3: Junta les columnes Fills i No Fills en una sola columna, utilitzant el mètode .apply() i definint una funció que resolgui el problema. 
# La columna nova ha de dir-se "Fills" i prendre els valors "Sí" o "No".


# Definim una funció per aplicar a cada fila
def determinar_fills(fila):
    if fila["Fills"] is True:
        return "Sí"
    elif fila["No Fills"] is True:
        return "No"

# Creem la nova columna unificada amb apply
df["Fills"] = df.apply(determinar_fills, axis=1)

# Eliminem la columna "No Fills"
df = df.drop(columns=["No Fills"])

df
```

```python
# EXERCICI 4: 
# Crea una taula resum que permeti veure el sou mig, medià, mínim i màxim per Gènere.
# Ordena la taula en funció del sou mig.

df["Salari mensual"] = (df["Salari mensual"].str.replace("€", "").str.replace(".", ""))
# Elimina els punts de milers i els símbols d'euro
df["Salari mensual"] = pd.to_numeric(df["Salari mensual"], errors="coerce")  # converteix el valor de salari a float

# Taula resum per Gènere
resum_salari = df.groupby("Gènere")["Salari mensual"].agg(
    Salari_mig = "mean",
    Salari_medià = "median",
    Salari_mínim = "min",
    Salari_màxim = "max").sort_values(by = "Salari_mig", ascending=False).round(2) # ordenem en funció del sou mig

resum_salari
```

```python
# Nivell 2 - EXERCICI 1 
# Utilitzant el següent DataFrame, adjunta la columna "Increment" al dataframe del nivell anterior.

df_increment = pd.DataFrame({"Grup":["Grup A","Grup B","Grup C", "Grup D" ] , "Increment":
["5%","3,5%","2%","8%"]})

# Actualitza la columna salari en funció dels percentatges que s'adjunten. 
# No modifiquis manualment els increments, escriu codi Python per fer les conversions necessàries.

# Combinem els dataframes amb un merge. 
df = df.merge(df_increment, how="left", left_on="Grup Professional", right_on="Grup")

# Eliminem la columna "Grup" que prové de df_increment
df.drop(columns=["Grup"], inplace=True)

# Converteim l'increment a format decimal
df["Increment"] = df["Increment"].str.replace("%", "").str.replace(",", ".").astype(float) / 100

# Calculem el nou salari amb l'increment aplicat, actualitzant la columna.
df["Salari mensual"] = df["Salari mensual"] * (1 + df["Increment"])
df.drop(columns=["Increment"], inplace= True) # treiem la columna increment, una vegada aplicat a la columna salari mensual
df  
```

```python
# NIVELL 3 - EXERCICI 1: Crea una funció que prengui un dataframe com a paràmetre d'entrada.
# La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:
# un histograma/boxplot si la variable és numèrica
# unes barres dels valors més freqüents si és categòrica
# unes barres dels anys més freqüents si la dada està en format data.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Definim la funció pel dataframe que utilitzarem i per l'exportació dels arxius
def crea_i_exporta_grafics(df, carpeta="graficos_exportats"):
    os.makedirs(carpeta, exist_ok = True) # afegim la condició exist_ok = True pq no doni error si ja l'hem creat.

# Recorrem totes les columnes del dataframe, eliminant els possibles valors buits.
    for col in df.columns:
        plt.figure()
        col_data = df[col].dropna()

        # Si és una data, fa un gràfic de barres per any 
        if pd.api.types.is_datetime64_any_dtype(col_data):
            col_data.dt.year.value_counts().sort_index().plot(kind="bar") # passar dues variables a l'hora de fer el gràfic
            plt.title(f"Freqüència per any: {col}")

        # Si és numèrica, fa un histograma
        elif pd.api.types.is_numeric_dtype(col_data):
            sns.histplot(col_data, kde=True)
            plt.title(f"Histograma: {col}")

        # Si és objecte o categòrica, fa un barplot
        elif isinstance(col_data.dtype, pd.CategoricalDtype) or col_data.dtype == object:
            col_data.value_counts().plot(kind="bar")
            plt.title(f"Barplot: {col}")
        else:
            plt.close()
            continue

        plt.tight_layout()
        plt.savefig(f"{carpeta}/{col}.png")
        plt.close()

    print(f"Els gràfics s'han guardat a: {os.path.abspath(carpeta)}")

import seaborn as sns

# Carreguem el dataset
df_penguins = sns.load_dataset("penguins")

# Executem la funció
crea_i_exporta_grafics(df_penguins)
```

```python
# EXERCICI 2: Ens interessa visitar totes les ciutats principals d'Espanya recorrent la mínima distància possible.

import pandas as pd

# Llegeix el fitxer amb la columna de ciutats com a índex
dataframe = pd.read_excel(r"C:\Users\joanr\OneDrive\Escritorio\FAD\matriu_distancies.xlsx", index_col=0)

# Eliminem les ciutats que no es pot arribar en cotxe
dataframe = dataframe.drop(index=["Palma", "Las Palmas de Gran Canaria"], columns=["Palma", "Las Palmas de Gran Canaria"]).fillna(0)

# Definim la funció:
def calcular_ruta_mes_propera(dist_matriu, origen):
    ciutats = list(dist_matriu.index) # com hem establert les ciutats com a índex del dataframe, les podem llistar.
    visitades = [origen] # les ciutats visitades pasaran a una llista per no repetir-les (començant amb la d'origen)
    actual = origen # la ciutat en la que estem
    distancia_total = 0 # acumularà la distancia total de la ruta

    while len(visitades) < len(ciutats):
        ciutats_no_visitades = [c for c in ciutats if c not in visitades] # mentre no haguem visitat la ciutat, recorrem la llista
        propera_ciutat = min(ciutats_no_visitades, key=lambda c: dist_matriu.loc[actual, c]) # escollim la ciutat més propera partint de l'actual
        distancia_total += dist_matriu.loc[actual, propera_ciutat]
        visitades.append(propera_ciutat)
        actual = propera_ciutat # sumem la distancia total i afegim la ciutat a les visitades

    return visitades, distancia_total

# Crida de la funció
ruta, distancia = calcular_ruta_mes_propera(dataframe, "Barcelona")
print("Ruta:", ruta)
print("Distància total:", distancia, "km")

```
