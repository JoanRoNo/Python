```python
import pandas as pd

df = pd.read_excel(r"C:\Users\joanr\OneDrive\Escritorio\FAD\sprint10.xlsx")
print(type(df))  # hauria de ser <class 'pandas.core.frame.DataFrame'>

# Eliminem les dues primeres files (0 i 1)
df = df.drop(index=[0, 1]).reset_index(drop=True)

# Assignem la primera fila com a nova capçalera (columnes)
df.columns = df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)

# Eliminem les dues primeres columnes
df = df.drop(df.columns[[0]], axis=1)

# Mostrem el DataFrame final
df

# Ordena el DataFrame pel país d'origen. En cas d'empat, ordena pel nom de la ciutat.
df.sort_values(["País d'origen", "Ciutat"], ascending=(True, True), inplace= True)
df


```

```python
# Addicionalment, fes un print on comprovi que el DNI només té valors únics.
cuenta = df["DNI"].value_counts()
print(cuenta)

numero_unicos = df["DNI"].nunique()
print(f"Número de valores únicos en DNI: {numero_unicos}")
```

```python
# Crea una columna que sigui el nom complet.
df['Nom_Cognoms'] = df[["Nom", "Cognoms"]].agg(" ".join, axis=1)


# Crea una columna si la persona és nascuda a Espanya o no.
df["Espanyol"] = df["País d'origen"].apply(lambda x: "Sí" if x == "Espanya" else "No")

# Posa el DNI com a índex del DataFrame (noms de files).
df.set_index("DNI", inplace= True)

# Substitueix el nom de les columnes Dia de Naixement, Mes de Naixement i Any de Naixement per Dia, Mes i Any.
df.rename(columns={"Dia de Naixement": "Dia", "Mes de Naixement": "Mes", "Any de Naixement": "Any"}, inplace= True)

# Substitueix H per Home, D per Dona, A per Altres i NC per una dada faltant (nan/null/na).
df["Gènere"] = df["Gènere"].replace({"A": "Altres", "H": "Home", "D": "Dona", "NC": "Null"})
df
```

```python
# Junta les columnes Fills i No Fills en una sola columna, utilitzant el mètode .apply() i definint una funció que resolgui el problema. 
# La columna nova ha de dir-se "Fills" i prendre els valors "Sí" o "No"

df["Fills"] = df["Fills"].apply(lambda x: True if str(x).strip().lower() == "true" else None)
df["No Fills"] = df["No Fills"].apply(lambda x: True if str(x).strip().lower() == "true" else None)
df

```

```python


df["Salari mensual"] = (
    df["Salari mensual"]
    .astype(str)
    .str.replace(r"[^\d,\.]", "", regex=True)  # Elimina símbols no numèrics (€, espais, etc.)
    .str.replace(".", "", regex=False)         # Elimina punts de milers
    .str.replace(",", ".", regex=False)        # Substitueix la coma decimal per punt
)
df["Salari mensual"] = pd.to_numeric(df["Salari mensual"], errors="coerce")  # Converteix a float

# Taula resum per Gènere
resum_salari = df.groupby("Gènere")["Salari mensual"].agg(
    Salari_mig="mean",
    Salari_medià="median",
    Salari_mínim="min",
    Salari_màxim="max"
).sort_values(by="Salari_mig", ascending=False).round(2)

resum_salari

```

```python
# Crea una taula resum amb el salari mig per gènere (files) i país d'origen (columnes).
# Afegeix-hi les mitjanes als marges de la taula.
# (EXTRA): Aplica format condicional a la taula per veure en un color més intens els valors més elevats

taula_resum = pd.pivot_table(
    df,
    values="Salari mensual",
    index="Gènere",
    columns="País d'origen",
    aggfunc="mean",
    margins=True,              
    margins_name="Mitjana"
).round(2)
taula_resum

```

```python
# Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. 
# Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.
# Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

df["Data de Naixement"] = df[["Dia", "Mes", "Any"]].astype(str).agg("-".join, axis=1)

from datetime import datetime
df["Data de Naixement"] = pd.to_datetime(df["Data de Naixement"], format="%d-%m-%Y", errors="coerce")

# Pas 2: calcular edat
avui = pd.to_datetime("today")

df["Edat"] = df["Data de Naixement"].apply(
    lambda data: avui.year - data.year - ((avui.month, avui.day) < (data.month, data.day))
    if pd.notnull(data) else None
)
df  
```
