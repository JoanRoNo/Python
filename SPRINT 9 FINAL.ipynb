```python
# NIVELL 1 - EXERCICI 1 - Calculadora de l'índex de massa corporal

#  Funció per demanar un valor positiu amb validació de dades 
def demanar_valor_positiu(text):
    while True:
        try:
            valor = float(input(text))
            if valor <= 0:
                print("Error: el valor ha de ser positiu.")
            else:
                return valor
        except ValueError:
            print("Error: introdueix un número vàlid. Ex: 70.5")

# Funció per calcular l'IMC
def calcular_imc(pes, alcada):
    return pes / (alcada ** 2)

# Funció per classificar segons els intervals definits per la pàgina de l'enunciat
def classificar_imc(imc):
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc <= 24.9:
        return "Peso normal"
    elif 25 <= imc <= 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Funció principal que crida les funcions prèviament definides
def calcular_i_mostrar_imc():
    imcs = {"Bajo peso": [], "Peso normal": [], "Sobrepeso": [], "Obesidad": []}

    pes = demanar_valor_positiu("Quant peses en kg? ")
    alçada = demanar_valor_positiu("Quant medeixes en metres? ")
    imc = calcular_imc(pes, alçada)
    categoria = classificar_imc(imc)

    # Afegim l'IMC a la categoria corresponent
    imcs[categoria].append(imc)

    print(f"\nResultats ordenats per categoria per {alçada}m i {pes}kg")
    print(f"El teu IMC és de {round(imc, 2)}. Classificació: {categoria}")
    for clau in imcs:
        print(f"{clau}:", sorted(imcs[clau]))
""" Amb aquesta funció podrem calcular i classificar l'IMC segons la categoria que li correspon """
calcular_i_mostrar_imc()
```

```python
# EXERICI 2 - Conversor de temperatures

# Funció per convertir temperatures
def conversor_temperatura(valor, unitat):
    unitat = unitat.lower()
    conversions = {}

    if unitat == "c" or unitat == "celsius":
        conversions["fahrenheit"] = valor * 9/5 + 32
        conversions["kelvin"] = valor + 273.15

    elif unitat == "f" or unitat == "fahrenheit":
        celsius = (valor - 32) * 5/9
        conversions["celsius"] = celsius
        conversions["kelvin"] = celsius + 273.15

    elif unitat == "k" or unitat == "kelvin":
        celsius = valor - 273.15
        conversions["celsius"] = celsius
        conversions["fahrenheit"] = celsius * 9/5 + 32

    else:
        raise ValueError("Unitat no reconeguda. Usa 'C', 'F' o 'K'.")

    return conversions

# Entrada de l'usuari i impressió dels resultats
try:
    valor = float(input("Introdueix la temperatura: "))
    unitat = str(input("Introdueix la unitat (C, F o K): "))

    resultats = conversor_temperatura(valor, unitat)

    # Diccionari amb els símbols de cada unitat
    simbols = {
        "celsius": "°C",
        "fahrenheit": "°F",
        "kelvin": "K"}

    print(f"\nPer {valor} {unitat.upper()}, les conversions són:")

    for k, v in resultats.items():
        simbol = simbols.get(k, "")
        print(f"{k}: {v:.2f} {simbol}")

except (KeyError, ValueError):
    print("S'ha produït un error: o bé no has introduït un número o la unitat introduïda no està especificada")
```

```python
# Exercici 3 - Comptador de paraules d'un text + mitjana

import string

# Funció per netejar i preparar el text
def netejar_text(text):
    text = text.lower()  # passem a minúscules
    for signe in string.punctuation:
        text = text.replace(signe, "")  # eliminem puntuació
    return text

# Funció per comptar paraules
def comptar_paraules(text):
    net = netejar_text(text)
    paraules = net.split()
    comptador = {}

    for paraula in paraules:
        if paraula in comptador:
            comptador[paraula] += 1
        else:
            comptador[paraula] = 1

    return comptador, paraules

# EXTRA: Funció per calcular la llargada mitjana de les paraules
def llargada_mitjana(paraules):
    if not paraules: 
        return 0 # en el cas que no s'escrigui res, la mitjana donarà 0.
    suma = sum(len(p) for p in paraules)
    return round(suma / len(paraules), 2)

# Funció principal
"""
Aquesta és la funció que s’executa i demana l’entrada a l’usuari.
Calcula i mostra la freqüència de cada paraula i la llargada mitjana de les paraules.
"""
def analitzar_text():
    text = input("Escriu un text: ")
    if not text.strip():
        print("No s'ha introduït cap text.")
        return
    comptador, paraules = comptar_paraules(text)
    mitjana = llargada_mitjana(paraules)

    print("\nFreqüència de paraules:")
    for paraula, compte in comptador.items():
        print(f"{paraula}: {compte}")

    print(f"Llargada mitjana de les paraules: {mitjana}")


""" La frase introduïda ha estat: Hola amics, què tal, això és una prova amics. 
D'aquesta manera es comprova que els signes de puntuació no afecten a les paraules introduides
és a dir, no es té en compte ni majúsucules ni comes ni punts davant o darrere """
analitzar_text()
```

```python
# EXERCICI 4 - Diccionari invers (amb possibilitat de duplicats)

# Funció per crear un diccionari a partir de n parells clau-valor
def crear_diccionari(n):
    diccionari = {}
    for i in range(n):
        clau = input(f"Introdueix la clau #{i+1}: ")
        valor = input(f"Introdueix el valor associat a "{clau}": ")
        diccionari[clau] = valor
    return diccionari

# Funció per invertir el diccionari original amb la possibilitat de duplicats.
def invertir_diccionari_amb_llistes(diccionari_original):
    diccionari_invers = {}
    for clau, valor in diccionari_original.items():
        diccionari_invers.setdefault(valor, []).append(clau)
    return diccionari_invers

n = int(input("Quants parells clau-valor vols introduir? "))
diccionari = crear_diccionari(n)

print("Diccionari original:")
print(diccionari)

diccionari_invers = invertir_diccionari_amb_llistes(diccionari)

print("\nDiccionari invers (amb valors duplicats agrupats):")
print(diccionari_invers)
```

```python
# NIVELL 2 - EXERCICI 1 - Comptador i endreçador de paraules d'un text.

import string

# Obrim el fitxer i llegim el contingut
with open("C:/Users/joanr/OneDrive/Escritorio/FAD/tu me quieres blanca.txt", "r", encoding= "utf-8") as poema:
    text = poema.read()

# Eliminem signes de puntuació i passem tot a minúscules
text = text.lower().translate(str.maketrans("", "", string.punctuation))

# Separem en paraules
llista_paraules = text.split()

# Diccionari principal on s'emagatzemaran les paraules i el nombre de vegades que apareixen.
diccionari_poema = {}

for paraula in llista_paraules:
    if not paraula[0].isalpha():
        continue  # Salta paraules que no comencin per lletra (números o símbols)

    inicial = paraula[0]

    if inicial not in diccionari_poema:
        diccionari_poema[inicial] = {}

    if paraula in diccionari_poema[inicial]:
        diccionari_poema[inicial][paraula] += 1
    else:
        diccionari_poema[inicial][paraula] = 1

# Funció principal
def mostrar_diccionari_ordenat(diccionari):
    """Mostra les paraules agrupades per lletra inicial i ordenades alfabèticament. """
    for lletra in sorted(diccionari):
        print(f"Paraules que comencen amb {lletra} i les vegades que apareixen:")
        for paraula in sorted(diccionari[lletra]):
            print(f"{paraula}: {diccionari[lletra][paraula]}")

mostrar_diccionari_ordenat(diccionari_poema)
```

```python
# EXERCICI 2 - Conversió de tipus de dades

# Funció per separar els elements convertibles a float dels que no
def separar_convertibles(llista):
    """
    Aquesta funció recorre la llista de l'enunciat que pot contenir elements simples o estructures com tuples o llistes, i separa:
    - Els elements que es poden convertir a float.
    - Els que no es poden convertir (són ignorats o generen error).
    Retorna dues llistes: convertibles i no convertibles.
    """
    convertibles = []
    no_convertibles = []

    for element in llista:
        # Si l’element és una llista o una tupla, la processem recursivament
        if isinstance(element, (list, tuple)):
            sub_convertibles, sub_no_convertibles = separar_convertibles(element)
            convertibles.extend(sub_convertibles)
            no_convertibles.extend(sub_no_convertibles)
        else:
            # Intentem convertir a float
            try:
                valor = float(element)
                convertibles.append(valor)
            except (ValueError, TypeError):
                no_convertibles.append(element)

    return convertibles, no_convertibles

# Execució amb l'exemple de l'enunciat
dades = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

# Processem les dades
floats, errors = separar_convertibles(dades)

# Mostrem resultats

""" veiem que els els valors en forma d'string o amb símbols no s'han pogut convertir a float"""
print("Valors convertits a float:")
print(floats)

print("\nValors NO convertibles:")
print(errors)
```

```python
# NIVELL 3 - EXERCICI 1 - Generador de Contrasenyes

import numpy as np
import string
import random
import pyperclip  # Per copiar al porta-retalls (EXTRA)

# Funció per crear una contrasenya aleatòria
def crear_contrasenya(longitud, majuscules=True, minuscules=True, numeros=True, signes=False, copiar=False):
    """
    Genera una contrasenya aleatòria segons els paràmetres indicats:
    - longitud (int): nombre total de caràcters
    - majuscules, minuscules, numeros, signes (bool): tipus de caràcters permesos
    - copiar (bool): si True, copia la contrasenya al porta-retalls (EXTRA)
    """

    # Definim els conjunts de caràcters disponibles segons paràmetres
    caracters = ""
    grups = []  # Ens servirà per garantir almenys un de cada

    if majuscules:
        caracters += string.ascii_uppercase
        grups.append(string.ascii_uppercase)

    if minuscules:
        caracters += string.ascii_lowercase
        grups.append(string.ascii_lowercase)

    if numeros:
        caracters += string.digits
        grups.append(string.digits)

    if signes:
        caracters += "!@#$%&*?,.-_+="
        grups.append("!@#$%&*?,.-_+=")

    if longitud < len(grups):
        raise ValueError(f"La longitud mínima ha de ser com a mínim {len(grups)} per incloure tots els tipus seleccionats.")

    # Generem la contrasenya:
    contrasenya = []

    # Garantim que hi hagi almenys un caràcter de cada grup seleccionat
    for grup in grups:
        contrasenya.append(np.random.choice(list(grup)))

    # Generem la resta de caràcters de forma totalment aleatòria
    resta = longitud - len(contrasenya)
    contrasenya += list(np.random.choice(list(caracters), size=resta))

    # Barregem la contrasenya per evitar patrons
    random.shuffle(contrasenya)

    # La unim com a string final
    contrasenya_final = "".join(contrasenya)

    # EXTRA: copiar al porta-retalls
    if copiar:
        try:
            pyperclip.copy(contrasenya_final)
            print("Contrasenya copiada al porta-retalls")
        except Exception as e:
            print(f"No s'ha pogut copiar al porta-retalls: {e}")

    return contrasenya_final
pwd = crear_contrasenya(10, True, True, True, True, copiar=True) # podem modificar la longitud de la contrasenya, canviant el 1r paràmetre.
print("Contrasenya generada:", pwd)
```

```python
# EXERCICI 2 - Processament de dades simple

from collections import defaultdict

# Inicialitzem diccionaris
gols_fets = defaultdict(int)
gols_rebuts = defaultdict(int)
punts = defaultdict(int)

# Ruta del fitxer
fitxer = r"C:\Users\joanr\OneDrive\Escritorio\FAD\historic partits.txt"

# Llegim el fitxer línia a línia
with open(fitxer, "r", encoding="utf-8") as f:
    for linia in f:
        try:
            equip1, resultat, equip2 = linia.strip().split("\t")
            gols1, gols2 = map(int, resultat.split("-"))

            # Gols fets i rebuts
            gols_fets[equip1] += gols1
            gols_fets[equip2] += gols2
            gols_rebuts[equip1] += gols2
            gols_rebuts[equip2] += gols1

            # Assignació de punts
            if gols1 > gols2:
                punts[equip1] += 3
            elif gols1 < gols2:
                punts[equip2] += 3
            else:
                punts[equip1] += 1
                punts[equip2] += 1
        except ValueError:
            continue  # Ignorem línies mal formades

# Trobar equip més golejador i més golejat
equip_mes_golejador = max(gols_fets, key=gols_fets.get)
equip_mes_golejat = max(gols_rebuts, key=gols_rebuts.get)

# Mostrar resultats i formateig de la taula resultant.
print("Equip més golejador:", equip_mes_golejador, "-", gols_fets[equip_mes_golejador], "gols")
print("Equip més golejat:", equip_mes_golejat, "-", gols_rebuts[equip_mes_golejat], "gols")

print("\nClassificació global:")

print("\n{:<25} {:>10} {:>10} {:>10}".format("Equip", "GF", "GR", "Punts"))
for equip in sorted(punts, key=punts.get, reverse=True):
    print("{:<25} {:>10} {:>10} {:>10}".format(equip, gols_fets[equip], gols_rebuts[equip], punts[equip]))
```
