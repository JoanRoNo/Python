```python
# NIVELL 1 - EXERCICI 1:  Consulta l‚ÄôAPI p√∫blica JSONPlaceholder utilitzant el m√®tode GET per obtenir:

# Llista de publicacions (/posts)
import requests
posts = requests.get("https://jsonplaceholder.typicode.com/posts")
if posts.status_code == 200:
    print("posts descarregats correctament:")
    print(posts.json())
else:
    print("error en posts", posts.status_code)


# Llista d‚Äôusuaris (/users)
users = requests.get("https://jsonplaceholder.typicode.com/users")
if users.status_code == 200:
    print("usuaris descarregats correctament")
    print(users.json())
else:
    print("error en usuaris", users.status_code)

# Llista de tasques (/todos)
tasques = requests.get("https://jsonplaceholder.typicode.com/todos")
if users.status_code == 200:
    print("tasques descarregades correctament")
    print(tasques.json())
else:
    print("error en usuaris", tasques.status_code)

# importem requests i agafem la llista de publicacions, usuaris i tasques amb get,.
# printarem que s'han descarregat correctament els posts i el contingut en el cas d'establir correctament la connexi√≥


# EXERCICI 2 - Mostra per pantalla:
# La quantitat total de cada recurs.
quant_posts = len(posts.json())
print(f"Hi ha un total de {quant_posts} publicacions")

quant_users= len(users.json())
print(f"Hi ha un total de {quant_users} usuaris")

quant_tasques = len(tasques.json())
print(f"Hi ha un total de {quant_tasques} tasques")

# mitjan√ßant len(variable.json()) podem obtenir la quantitat total de cada recurs.



# EXERCICI 3 - Fes una petici√≥ a una publicaci√≥ inexistent per obtenir un error 404 i mostra el codi d'estat rebut.
fake_post = requests.get("https://jsonplaceholder.typicode.com/posts/99999")

print("Publicaci√≥ inexistent:")
print("Codi d‚Äôestat rebut:", fake_post.status_code)

# Buscant amb requests.get una publicaci√≥ amb un codi inexistent (99999), quan li demanem el codi d'estatus ens tornar√† un error conforme no s'ha trobat l'url proporcionada.

# EXERCICI 4 - Fes una petici√≥ POST per crear una nova publicaci√≥ fict√≠cia. Inclou el t√≠tol, el cos del missatge i un userId.
resposta = requests.post("https://jsonplaceholder.typicode.com/posts", 
                         data={"userId": 101, 
                               "id": 101, 
                               "title": "Hola", 
                               "body": "Hola que tal amics, aquesta es la meva primera publi"})

# Amb la petici√≥ POST podem crear una publicaci√≥, afegint els camps necessaris dins de publicacions (id 101 perqu√® hi ha 100 publicacions, afegim una m√©s )

# Mostra la resposta JSON
print(f" Resposta JSON {resposta.json()}")

# Mostra el codi d'estat.
print(f"codi d'estat: {resposta.status_code}")


# EXERCICI 5 - Fes una petici√≥ PATCH per modificar parcialment una publicaci√≥ existent.
modificacio = {"title": "T√≠tol modificat"}
resposta = requests.patch("https://jsonplaceholder.typicode.com/posts/1", json=modificacio)
# amb la condici√≥ patch podem canviar qualsevol camp de les publicacions, en aquest cas es modifica el t√≠tol.

# Mostra la resposta JSON.
print(f"Resposta JSON: {resposta.json()}")

# Mostra el codi d'estat.
print(f"Codi destat: {resposta.status_code}")


# EXERCICI 6 - Fes una petici√≥ DELETE sobre una publicaci√≥.
resposta = requests.delete("https://jsonplaceholder.typicode.com/posts/2")
# en aquest cas, amb la petici√≥ delete eliminem el post q volem.

# Mostra la resposta JSON.
print(f"Resposta JSON: {resposta.json()}")

# Mostra el codi d'estat.
print(f"Codi destat: {resposta.status_code}")

# quan mostrem la reposta JSON veurem un diccionari buit, ja que la publicaci√≥ que estem sol¬∑licitant ja no existeix.
```

```python
# EXERCICI 2: Dels resultats obtinguts, selecciona‚Äôn un que tingui recursos disponibles en CSV o JSON.

url_search = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search?q=aigua"
resposta = requests.get(url_search).json()

print("Nombre de datasets trobats:", resposta["result"]["count"])

for ds in resposta["result"]["results"]:
    url_show = f"https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_show?id={ds['id']}"
    r = requests.get(url_show).json()
    recursos = r["result"]["resources"]
    for rec in recursos:
        if rec["format"].lower() in ["csv", "json"]:
            print("\nDataset:", ds["title"])
            print("-", rec["format"], rec["url"])
            print("resource_id:", rec["id"])   

# Busca datasets d‚Äôaigua i llista nom√©s aquells recursos descarregables en CSV o JSON, amb el seu resource_id.
# Nom√©s mostra la informaci√≥ √∫til d'aquests: t√≠tol, url i id (necessari per fer consultes)
```

```python
# EXERCICI 3: Amb el resource_id del recurs seleccionat, realitza una consulta mitjan√ßant datastore_search per: Recuperar almenys 100 registres.

import requests

# Exemple de resource_id d'un dataset d'aigua 
resource_id = "0e3e679f-2255-4f8e-ac05-983178fa0a99" # seleccionem el de fonts ornamentals d'aigua de barcelona

# Endpoint datastore_search amb limit=100
url = f"https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search?resource_id={resource_id}&limit=100"

resposta = requests.get(url).json()

if resposta.get("success"):
    dades = resposta["result"]["records"]
    print("Nombre de registres recuperats:", len(dades))
    
    # Mostrem els 3 primers registres per exemple
    for row in dades[:3]:
        print(row)
else:
    print("Error:", resposta)
```

```python
# EXERCICI 4: Converteix els resultats a un DataFrame.

import pandas as pd

# Fem la crida al datastore_search (100 registres com a m√≠nim)
url = f"https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search?resource_id={resource_id}&limit=100"
resposta = requests.get(url).json()

# Extraiem els registres
records = resposta["result"]["records"]

# Convertim a DataFrame
df = pd.DataFrame(records)

# Mostrem el dataframe
df
```

```python
# EXERCICI 5:  Desa el DataFrame en un fitxer .csv.

# Desa el DataFrame en un CSV (sense index)
df.to_csv("dades_aigua.csv", index=False, encoding="utf-8")

print("Fitxer 'dades_aigua.csv' creat correctament!")
```
